# Задание 1

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston 
```

```python
boston = load_boston() #загрузка датасета в переменную boston 
```

```python
boston.keys() # Содержимое датасета
```

```python
data = boston.data
data.shape
```

```python
data
```

```python
target = boston.target 
target
```

```python
feature_names = boston.feature_names
feature_names
```

```python
for line in boston.DESCR.split('/n'):
    print(line)
```

```python
X = pd.DataFrame(data, columns = feature_names)
X.head()
```

```python
y = pd.DataFrame(target, columns = ['price'])
y.head()
```

```python
from sklearn.model_selection import train_test_split # import function train_test
```

### Parse the data into training and test 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

### Building a linear regression model

```python
from sklearn.linear_model import LinearRegression 
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train) #Model training 
```

```python
y_pred = lr.predict(X_test) #Prediction 
```

```python
check_test = pd.DataFrame({'y_test': y_test['price'],
                           'y_pred': y_pred.flatten()},
                          columns = ['y_test', 'y_pred'])
```

```python
check_test.head(10)
```

### R2

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
