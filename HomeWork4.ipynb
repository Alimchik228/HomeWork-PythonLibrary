# Задание 2


```python
t = np.linspace(0, 10, 51)
```

```python
f = np.cos(t)
```

```python
title_font = {
    "fontsize": 16,
    "family" : "serif",  
}

plt.plot(t, f, color = "green")
plt.title("График f(t)", fontdict = title_font)
plt.xlabel("Значения t", fontdict = title_font)
plt.ylabel("Значения f", fontdict = title_font)
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```

# Задание 3

```python

x = np.linspace(-3, 3, 51)
```

```python
y1 = x**2
y2 = 2 * x + 0.5 
y3 = -3 * x -1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2)
fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace = 0.3, hspace = 0.3)
ax[0][0].plot(x, y1)
ax[0][0].set_title("График y1")
ax[0][0].set_xlim([-5, 5])
ax[0][1].plot(x, y2)
ax[0][1].set_title("График y2")
ax[1][0].plot(x, y3)
ax[1][0].set_title("График y3")
ax[1][1].plot(x, y4)
ax[1][1].set_title("График y4")
```
